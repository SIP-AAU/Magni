# `imaging-preprocessing`: Show examples of how to remove tilt from an image

This example primarily shows the usage of the following functions:
- `magni.imaging.preprocessing.detilt`

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

import magni
```

```python
# Load example
img = magni.afm.io.read_mi_file('example.mi').get_buffer('Topography')[0].get_data()
img = magni.imaging.visualisation.stretch_image(img, 1)

# Show image
magni.utils.plotting.setup_matplotlib({'figure': {'figsize': (20, 10)}})
_ = magni.imaging.visualisation.imshow(img)
```

Remove tilt by fitting and subtracting a plane

```python
# Remove plane tilt
plane_detilt_img, plane_tilt = magni.imaging.preprocessing.detilt(img, mode='plane_flatten', return_tilt=True)

# Show detilted image and fitted plane
fig, axes = plt.subplots(1, 2)
a_i = magni.imaging.visualisation.imshow(plane_detilt_img, ax=axes[0])
plt.colorbar(a_i, ax=axes[0], shrink=0.5)
axes[0].set_title('Detilted image')

a_t = magni.imaging.visualisation.imshow(plane_tilt, ax=axes[1])
plt.colorbar(a_t, ax=axes[1], shrink=0.5)
_ = axes[1].set_title('Estimated plane tilt')
```

Remove tilt by fitting and subtracting polynomials of degree 1

```python
# Remove line tilt (first order)
line1_detilt_img, line1_tilt = magni.imaging.preprocessing.detilt(img, mode='line_flatten', degree=1, return_tilt=True)

# Show detilted image and fitted lines
fig, axes = plt.subplots(1, 2)
a_i = magni.imaging.visualisation.imshow(line1_detilt_img, ax=axes[0])
plt.colorbar(a_i, ax=axes[0], shrink=0.5)
axes[0].set_title('Detilted image')

a_t = magni.imaging.visualisation.imshow(line1_tilt, ax=axes[1])
plt.colorbar(a_t, ax=axes[1], shrink=0.5)
_ = axes[1].set_title('Estimated line tilt')
```

Remove tilt by fitting and subtracting polynomials of degree 2

```python
# Remove line tilt (second order)
line2_detilt_img, line2_tilt = magni.imaging.preprocessing.detilt(img, mode='line_flatten', degree=2, return_tilt=True)

# Show detilted image and fitted lines
fig, axes = plt.subplots(1, 2)
a_i = magni.imaging.visualisation.imshow(line2_detilt_img, ax=axes[0])
plt.colorbar(a_i, ax=axes[0], shrink=0.5)
axes[0].set_title('Detilted image')

a_t = magni.imaging.visualisation.imshow(line2_tilt, ax=axes[1])
plt.colorbar(a_t, ax=axes[1], shrink=0.5)
_ = axes[1].set_title('Estimated line tilt')
```

Remove tilt by subtracting a plane fitted to a masked part of the image

```python
# Create spiral sample path
h = img.shape[0]; w = img.shape[1]; scan_length = 8000.0; num_points = 10000;
image_coords = magni.imaging.measurements.spiral_sample_image(h, w, scan_length, num_points)
unique_pixels = magni.imaging.measurements.unique_pixels(image_coords)

# Extract pixels masked by the spiral
mask = np.zeros_like(img, dtype=np.bool)
mask[unique_pixels[:, 1], unique_pixels[:, 0]] = True
masked_img = np.ma.array(img, mask=np.invert(mask))

# Show mask
fig, axes = plt.subplots(1, 2)
magni.imaging.visualisation.imshow(mask, ax=axes[0], cmap='gray')
magni.imaging.visualisation.imshow(masked_img, ax=axes[1])

# Remove plane tilt from image masked with spiral
plane_mask_detilt_img, plane_mask_tilt = magni.imaging.preprocessing.detilt(img, mask=mask, mode='plane_flatten', return_tilt=True)

# Show detilted image and fitted plane
fig, axes = plt.subplots(1, 2)
a_i = magni.imaging.visualisation.imshow(plane_mask_detilt_img, ax=axes[0])
plt.colorbar(a_i, ax=axes[0], shrink=0.5)
axes[0].set_title('Detilted image')

a_t = magni.imaging.visualisation.imshow(plane_mask_tilt, ax=axes[1])
plt.colorbar(a_t, ax=axes[1], shrink=0.5)
_ = axes[1].set_title('Estimated plane tilt')
```
